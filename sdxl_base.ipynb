```python
import ipywidgets as widgets
from PIL import Image
import torch
from diffusers import StableDiffusionXLPipeline, LMSDiscreteScheduler, DDIMScheduler
import random
import uuid
import re
import io

# Load model
model_path ="RealVisXLV5.0.safetensors"
model_path ="sd_xl_base_1.0.safetensors"
lora_path ="aodaipony.safetensors"
loras_path1 ="milf.safetensors"
lora_path ="p.safetensors"
pipe = StableDiffusionXLPipeline.from_single_file("p.safetensors", torch_dtype=torch.float16)
pipe.to("cuda")
#load lora
pipe.load_lora_weights(lora_path)
pipe.fuse_lora(lora_scale=0.5)  # Scale từ 0.0 đến 1.0
#pipe.load_lora_weights({
#    "aodai": "Aodaixl.safetensors",
#    "cukcold": "cukcold.safetensors"
#})
#pipe.fuse_lora(lora_scale={
  #  "aodai": 0.8,
 #   "cukcold": 0.5
#})

pipe.enable_attention_slicing()
pipe.enable_xformers_memory_efficient_attention()

# Set scheduler - Using LMSDiscreteScheduler or DDIMScheduler
#scheduler = LMSDiscreteScheduler.from_config(pipe.scheduler.config)
# Alternatively, you can use DDIM Scheduler
# scheduler = DDIMScheduler.from_config(pipe.scheduler.config)
from diffusers import DDIMScheduler
scheduler = DDIMScheduler.from_config(pipe.scheduler.config)
pipe.scheduler = scheduler


pipe.scheduler = scheduler

def sanitize_filename(prompt):
    return re.sub(r'[^a-zA-Z0-9]', '_', prompt)[:50]

def get_random_suffix():
    return str(uuid.uuid4().hex[:8])

negative_prompt = "Negative(hands), text, error, cropped, (worst quality:1.2), (low quality:1.2), normal quality, (jpeg artifacts:1.3), signature, watermark, username, blurry, artist name, monochrome, sketch, censorship, censor, (copyright:1.2), extra legs,(forehead mark),(depth of field),(emotionless),(penis), noises, dusts, scratches, multicolor, patches, extra hands, wrong adam apple, dark skin, painting, earring, scratches, stains, old photo damages, bad quality, blurry, (beard:2), (facial hair:2)"

width, height = 768, 1024

import os

save_dir = "/content/drive/MyDrive/Myxxx"
os.makedirs(save_dir, exist_ok=True)

def generate_image(prompt, num_steps, guidance_scale, denoise_strength, init_image=None):
    if init_image:
        image = pipe(prompt, negative_prompt=negative_prompt, init_image=init_image,
                     num_inference_steps=num_steps, guidance_scale=guidance_scale,
                     denoise_strength=denoise_strength, height=height, width=width).images[0]
    else:
        image = pipe(prompt, negative_prompt=negative_prompt,
                     num_inference_steps=num_steps, guidance_scale=guidance_scale,
                     denoise_strength=denoise_strength, height=height, width=width).images[0]

    filename = f"output_{sanitize_filename(prompt)}_{get_random_suffix()}.png"
    full_path = os.path.join(save_dir, filename)
    image.save(full_path)

    return image, full_path


prompt_widget = widgets.Text(value="Astronaut in a jungle, cold color palette, muted colors, detailed, 8k", description="Prompt:", style={'description_width': 'initial'})
num_steps_widget = widgets.IntSlider(value=30, min=3, max=100, step=1, description="Steps:", style={'description_width': 'initial'})
guidance_scale_widget = widgets.FloatSlider(value=7.5, min=1.0, max=20.0, step=0.1, description="Guidance Scale:", style={'description_width': 'initial'})
denoise_strength_widget = widgets.FloatSlider(value=0.5, min=0.0, max=1.0, step=0.1, description="Denoise Strength:", style={'description_width': 'initial'})

output_widget = widgets.Output()
thumbnail_widget = widgets.Output()

def create_upload_widget():
    return widgets.FileUpload(accept="image/*", multiple=False)

upload_widget = create_upload_widget()
uploaded_image = None

def on_upload_change(change):
    global uploaded_image
    with thumbnail_widget:
        thumbnail_widget.clear_output()

        if upload_widget.value:
            key = next(iter(upload_widget.value))
            image_data = upload_widget.value[key]['content']
            uploaded_image = Image.open(io.BytesIO(image_data)).convert("RGB")

            thumbnail = uploaded_image.copy()
            thumbnail.thumbnail((150, 150))
            display(thumbnail)

    # Reset upload widget để xóa số nhảy
    reset_upload_widget()

def reset_upload_widget():
    global upload_widget
    upload_widget_box.children = [create_upload_widget()]
    upload_widget_box.children[0].observe(on_upload_change, names='value')

def remove_uploaded_image(b):
    global uploaded_image
    uploaded_image = None
    with thumbnail_widget:
        thumbnail_widget.clear_output()

    reset_upload_widget()

remove_button = widgets.Button(description="Remove Image", button_style='danger')
remove_button.on_click(remove_uploaded_image)

upload_widget_box = widgets.VBox([upload_widget])
image_area = widgets.VBox([upload_widget_box, thumbnail_widget, remove_button])

upload_widget.observe(on_upload_change, names='value')

def update_image(prompt, num_steps, guidance_scale, denoise_strength):
    with output_widget:
        output_widget.clear_output(wait=True)

        if uploaded_image:
            image, filename = generate_image(prompt, num_steps, guidance_scale, denoise_strength, init_image=uploaded_image)
        else:
            image, filename = generate_image(prompt, num_steps, guidance_scale, denoise_strength)

        display(image)
        print(f"Saved as: {filename}")

generate_button = widgets.Button(description="Generate Image", button_style='success')
generate_button.on_click(lambda b: update_image(prompt_widget.value, num_steps_widget.value, guidance_scale_widget.value, denoise_strength_widget.value))

control_panel = widgets.VBox([prompt_widget, num_steps_widget, guidance_scale_widget, denoise_strength_widget, generate_button])
image_display = widgets.VBox([image_area, output_widget])

ui_layout = widgets.HBox([control_panel, image_display])

display(ui_layout)
```
